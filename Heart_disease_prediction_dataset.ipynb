Naive bayes

```python
from sklearn.model_selection import train_test_split 
from sklearn.naive_bayes import GaussianNB

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
Gmodel=GaussianNB() 

print('training input sample',x_train.shape)
print('training output label',y_train.shape)

print('testing input sample',x_test.shape)
print('testing output label',y_test.shape)
```

```python
Gmodel.fit(x_train,y_train) # implementing Gaussian module to find the pattern

train_Gpred=Gmodel.predict(x_train) 

test_Gpred=Gmodel.predict(x_test) 
```

```python
test_Gpred
```

```python
train_Gpred
```

```python
train_acc_gau=np.mean(train_Gpred==y_train)

test_acc_gau=np.mean(test_Gpred==y_test)

print('gaussian naive bayes - training accuracy:',train_acc_gau*100,'%')
print('gaussian nb - testing time accuracy:',test_acc_gau*100,'%')
```

Gaussian performs better than all algorithm

```python
from sklearn.naive_bayes import MultinomialNB
Mmodel=MultinomialNB()  # multinomial NB implementing
Mmodel.fit(x_train,y_train)

train_Mpred=Mmodel.predict(x_train)
test_Mpred=Mmodel.predict(x_test)
```

```python
train_acc_multi=np.mean(train_Mpred==y_train)
test_acc_multi=np.mean(test_Mpred==y_test)
print('multinomial naive bayes - training accuracy:',train_acc_multi*100,'%')
print('multinomial nb - testing time accuracy:',test_acc_multi*100,'%')
```

Random Forest 

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train=ss.fit_transform(x_train) # xtrain = training input samples
x_test=ss.transform(x_test) # xtest - testing input samples
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.ensemble import RandomForestClassifier

clfr=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)

clfr.fit(x_train,y_train)
```

```python
clfr1=RandomForestClassifier(n_estimators=10,criterion='gini',random_state=0)

clfr1.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

ypre=clfr1.predict(x_test)# entropy ypre calculation

ypre1=clfr1.predict(x_test)# gini ypre calculation
```

```python
print('entropy Accuracy Score:')
accuracy_score(y_test,ypre)*100
```

```python
print('gini Accuracy Score:')
accuracy_score(y_test,ypre1)*100
```
